```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv( './2012_subject_pool_partial.csv', index_col=0 )
df
```

```python
df.columns
```

```python
df['marstat']
```

```python
for data_key in df.columns:
  
  specific_columns_to_plot = [ 'birthyr', 'gender', 'educ' ]
  pre_to_plots = [ 'policy', 'pk' ]
  
  plot_condition = ( data_key.split( '_' )[0] in pre_to_plots ) or ( data_key in specific_columns_to_plot )
  
  if plot_condition:

    data = df[data_key]

    # Make the histogram itself
    edges = np.arange( data.min()-0.5, data.max() + 0.5, 1. )
    hist, edges = np.histogram(data, bins=edges )

    fig = plt.figure( figsize=(6,2), facecolor='white' )
    ax = plt.gca()

    # Inserting a 0 at the beginning allows plotting a numpy histogram with a step plot
    ax.step( edges, np.insert(hist, 0, 0.), color='black', linewidth=3.5 )

    ax.set_title( data_key, fontsize=16, )
```
